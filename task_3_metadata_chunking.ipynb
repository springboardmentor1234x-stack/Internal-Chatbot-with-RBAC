```python
import os
import uuid
import json

file_path = "/content/market_report_q4_2024.md"

with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

def make_chunks(text, chunk_size=350, num_chunks=3):
    chunks = []
    start = 0
    for i in range(1, num_chunks + 1):
        end = start + chunk_size
        chunk_text = text[start:end]
        metadata = {
            "chunk_id": f"chunk_{i}",
            "source_file": os.path.basename(file_path),
            "start_index": start,
            "end_index": end,
            "uuid": str(uuid.uuid4())
        }
        chunks.append({"text": chunk_text, "metadata": metadata})
        start = end
    return chunks

chunks = make_chunks(text, chunk_size=350, num_chunks=3)

for c in chunks:
    print("\n------------------------------")
    print("CHUNK TEXT:\n")
    print(c["text"])
    print("\nMETADATA:\n")
    print(json.dumps(c["metadata"], indent=4))
    print("------------------------------\n")

output_path = "/content/chunks_with_metadata.json"

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(chunks, f, indent=4, ensure_ascii=False)

print(f"Saved output → {output_path}")
```
