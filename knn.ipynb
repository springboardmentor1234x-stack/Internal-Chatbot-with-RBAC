```python
file_path = "/content/hr_data_cleaned.txt"

with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

rows = [line.strip() for line in lines[1:] if line.strip()]

documents = []
for row in rows:
    documents.append(" ".join(row.split()))

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
embeddings = vectorizer.fit_transform(documents).toarray()

import numpy as np

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (
        np.linalg.norm(vec1) * np.linalg.norm(vec2)
    )

def knn_search(query, k=3):
    query_vec = vectorizer.transform([query]).toarray()[0]
    scores = []
    for idx, doc_vec in enumerate(embeddings):
        score = cosine_similarity(query_vec, doc_vec)
        scores.append((idx, score))
    scores.sort(key=lambda x: x[1], reverse=True)
    return scores[:k]

query = input("Enter your query: ")
k = 3

results = knn_search(query, k)

print(f"\nTop {k} Results:\n")

for idx, score in results:
    print(f"Similarity: {score:.4f}")
    print(documents[idx])
    print("-" * 60)
```
