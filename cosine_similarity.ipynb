```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
chunks = [
    "Gross Margin: 60% (Up from 55% in 2023). Industry Benchmark: 55%. FinSolve Technologies is performing above average in operational efficiency and cost control.",
    "Net Margin: 12% (Consistent with 2023). Industry Benchmark: 10%. Stable net margin indicates control over direct and indirect costs.",
    "Days Sales Outstanding (DSO): 45 days. Industry Benchmark: 30 days. High DSO indicates receivables collection needs improvement."]
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embeddings = model.encode(chunks)
print("embeddings",embeddings)
similarity_matrix = cosine_similarity(embeddings)
print("\nCosine Similarity Matrix:\n")
print(similarity_matrix)
```
