```
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
model = SentenceTransformer('all-MiniLM-L6-v2')
sentences = [
    "I am Reethika A ",
    "Delhi is our capital city.",
    "The stock market saw a sharp decline.",
    "I enjoy learning about machine learning."
]
print("Sentences to embed:")
for i, s in enumerate(sentences):
    print(f"{i}: {s}")

embeddings = model.encode(sentences, convert_to_numpy=True)

print(f"embeddings matrix: {embeddings.shape}")
first_sentence_embedding = embeddings[0]
print(f"\nFirst 8 values of embedding for sentence 0 ('{sentences[0][:15]}...'):")
print(first_sentence_embedding[:8])
```
