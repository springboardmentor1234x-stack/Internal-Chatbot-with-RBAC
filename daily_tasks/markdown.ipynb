```python
import pandas as pd

# Raw markdown content (no preprocessing)
data = [
    {"line_no": 1, "raw_markdown": "# Introduction to Markdown"},
    {"line_no": 2, "raw_markdown": "Markdown is a lightweight markup language."},
    {"line_no": 3, "raw_markdown": "It is used to format plain text."},
    {"line_no": 4, "raw_markdown": "## Features"},
    {"line_no": 5, "raw_markdown": "- Easy to read"},
    {"line_no": 6, "raw_markdown": "- Easy to write"},
    {"line_no": 7, "raw_markdown": "- Widely used in GitHub"},
    {"line_no": 8, "raw_markdown": "## Example Code"},
    {"line_no": 9, "raw_markdown": "`print('Hello Markdown')`"},
    {"line_no": 10, "raw_markdown": "## Conclusion"},
    {"line_no": 11, "raw_markdown": "Markdown helps create structured documentation."}
]

df = pd.DataFrame(data)

# Save as CSV
csv_file = "markdown_raw_dataset.csv"
df.to_csv(csv_file, index=False)

print("CSV file created:", csv_file)
```

```python
import pandas as pd

df = pd.read_csv("markdown_raw_dataset.csv")
df.head()
```

```python
df["normalized_text"] = df["raw_markdown"].str.lower()
df.head()
#convert to lower case
```

```python
import re

df["normalized_text"] = df["normalized_text"].str.replace(r"[#`\-]", "", regex=True)
df.head()
#removing markdown symbols
```

```python
import string

df["normalized_text"] = df["normalized_text"].str.translate(
    str.maketrans("", "", string.punctuation)
)
df.head()
#removing punctuation
```

```python
df["normalized_text"] = df["normalized_text"].str.replace(r"\s+", " ", regex=True).str.strip()
df.head()
#remove extra spaces
```

```python
df = df[df["normalized_text"] != ""]
df.head()
#remove empty rows
```

```python
df.to_csv("normalized_markdown.csv", index=False)
#save normalized file
```
