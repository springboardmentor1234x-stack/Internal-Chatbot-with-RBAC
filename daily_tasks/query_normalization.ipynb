```python
queries = [
    "HOW can I apply for LEAVE???",
    "Show me   company   HOLIDAYS!!",
    "reset my PASSWORD pls",
    "Who is my TEAM manager???",
    "policy details!!!",
    "PAYROLL   info??",
    "system   LOGS!!!",
    "leave approval process???"
]
```

```python
#Normalization function
import re

def normalize_query(query):
    query = query.lower()
    query = re.sub(r'[^a-z\s]', '', query)
    query = re.sub(r'\s+', ' ', query).strip()
    return query
```

```python
# Applying normalization
normalized_queries = []

for q in queries:
    cleaned = normalize_query(q)
    normalized_queries.append(cleaned)
    print(f"Original: {q}")
    print(f"Normalized: {cleaned}\n")
```

```python
#vector database
#A vector database stores data as vectors and retrieves the most similar data using similarity search.
!pip install chromadb

from sentence_transformers import SentenceTransformer
import chromadb

model = SentenceTransformer("all-MiniLM-L6-v2")

client = chromadb.Client()
collection = client.get_or_create_collection(name="chatbot_queries")

# Define normalized_queries as a list of strings (example queries)
normalized_queries = [
    "What are the operating hours?",
    "How do I reset my password?",
    "Where can I find the nearest store?",
    "What is your return policy?",
    "How can I contact customer support?"
]

for i, query in enumerate(normalized_queries):
    embedding = model.encode(query).tolist()
    collection.add(
        ids=[str(i)],
        documents=[query],
        embeddings=[embedding]
    )
```

```python
#Simple Normalization Form (Text Normalization)

#Text normalization is the process of converting user queries into a standard, clean format so that the chatbot can understand them correctly.

#Normalization includes:

#->Converting text to lowercase

#->Removing punctuation

#->Removing extra spaces

#->Removing special characters

#->Removing stop words
```

```python
#Query Preprocessing Pipeline (Simple)

#1)The query preprocessing pipeline follows these steps:

#2)User enters a raw (messy) query

#3)Convert query to lowercase

#4)Remove punctuation and special characters

#5)Remove extra spaces

#6)Output cleaned query

#7)Convert cleaned query into vector embedding

#8)Store embedding in Vector Database

#9)Use vectors for similarity search
```

```python
#pipeline flow
'''User Query
   ↓
Lowercasing
   ↓
Remove punctuation & symbols
   ↓
Remove extra spaces
   ↓
Normalized Query
   ↓
Vector Embedding
   ↓
Vector Database'''
```
