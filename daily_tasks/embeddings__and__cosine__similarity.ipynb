```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Chunks
chunks = [
    "Artificial Intelligence focuses on developing systems that can perform tasks requiring human intelligence, such as learning and decision-making.",
    "Machine learning is a key part of artificial intelligence that enables systems to learn from data and improve their performance automatically.",
    "Artificial intelligence is widely applied in healthcare, finance, autonomous vehicles, and recommendation systems."
]

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Generate embeddings
embeddings = model.encode(chunks)

# Convert to numpy
embeddings = np.array(embeddings)

print("Embedding Vector Shape:", embeddings.shape)
```

```python
print("First 10 values of Chunk 1 embedding:")
print(embeddings[0][:10])
```

```python
similarity_matrix = cosine_similarity(embeddings)
print("Cosine Similarity Matrix:\n", similarity_matrix)
```

```python
#Which two chunks are most similar?
#Chunk 1 and Chunk 3 (similarity ≈ 0.7166)
#Why are Chunk 1 and Chunk 3 most similar?
#answer
#Because:

#Both strongly mention Artificial Intelligence

#Chunk 1 explains what AI is

#Chunk 3 explains where AI is applied

#They share common semantic concepts, so embeddings are closer
```

```python
#Did chunks from the same document have higher similarity?

#YES
#All chunks are from the same document and domain (AI)

#Similarities are moderate to high (0.56 – 0.71)

#This shows shared context .
```
